## 1. [4p] Rejection sampling for the density

We want to generate samples from the density

$$
f(x) = C\, x^{0.2}\,(1 - x)^{1.3}, \qquad 0 \le x \le 1,
$$

where the constant \(C\) normalizes the density. **We do not need to know the value of \(C\)** for rejection sampling.

---

### Step-by-step solution (by hand)

#### **Step 1 — Identify the unnormalized density**
Rejection sampling only needs the shape of the density.  
Define the unnormalized function:

$$
g(x) = x^{0.2} (1 - x)^{1.3}.
$$

Since \(C\) is unknown, we only ever evaluate \(g(x)\), not the true \(f(x)\).

---

#### **Step 2 — Choose a proposal distribution**
Because the support of \(f\) is \([0,1]\), the simplest choice is:

- Proposal:  
  $$
  X \sim \text{Uniform}(0,1)
  $$

- Proposal density:
  $$
  h(x) = 1.
  $$

---

#### **Step 3 — Find a constant \(M\) such that**
We need a constant \(M\) satisfying:

$$
f(x) \le M h(x)
\quad \Leftrightarrow \quad
g(x) \le M.
$$

So, we need the **maximum value** of \(g(x)\) on \([0,1]\):

$$
M = \max_{x \in [0,1]} x^{0.2}(1-x)^{1.3}.
$$

We do not need the exact analytic maximum—numerically evaluating it is fine.

---

#### **Step 4 — Rejection sampling algorithm**
For each desired sample:

1. Draw \(x \sim \text{Uniform}(0,1)\).  
2. Draw \(u \sim \text{Uniform}(0,1)\).  
3. Accept the sample if:
   $$
   u \le \frac{g(x)}{M}.
   $$  
4. If rejected, repeat.

This works because:

$$
\Pr(\text{accept } x) \propto g(x)
$$

and \(C\) cancels out automatically.

---

#### **Step 5 — Return \(n\) accepted samples**
Repeat until you have exactly \(n\) accepted draws.

---

### Final method summary
- Use uniform proposal on \([0,1]\).  
- Compute the maximum of \(g(x)\) numerically.  
- Use rejection sampling rule \(u < g(x)/M\).  
- Collect accepted samples.  

This produces samples from the intended density without needing the normalizing constant \(C\).
