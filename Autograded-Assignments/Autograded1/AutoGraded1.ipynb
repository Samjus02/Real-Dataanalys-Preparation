## Students passing exam (Sample exam problem)
Let's say we have an exam question which consists of $10$ yes/no questions. 
From past performance of similar students, a randomly chosen student will know the correct answer to $N \sim \text{binom}(10,6/10)$ questions. Furthermore, we assume that the student will guess the answer with equal probability to each question they don't know the answer to, i.e. given $N$ we define $Z \sim \text{binom}(10-N,1/2)$ as the number of correctly guessed answers. Define $Y = N + Z$, i.e., $Y$ represents the number of total correct answers.

We are interested in setting a deterministic threshold $T$, i.e., we would pass a student at threshold $T$ if $Y \geq T$. Here $T \in \{0,1,2,\ldots,10\}$.

1. [5p] For each threshold $T$, compute the probability that the student *knows* less than $5$ correct answers given that the student passed, i.e., $N < 5$. Put the answer in `problem11_probabilities` as a list.
2. [3p] What is the smallest value of $T$ such that if $Y \geq T$ then we are 90\% certain that $N \geq 5$?

## Students passing exam

We consider an exam with 10 yes/no questions.

- Let $N$ be the number of questions that a randomly chosen student **knows**.
- Let $Z$ be the number of correctly guessed answers (among the questions the student does *not* know).
- Let $Y = N + Z$ be the total number of correct answers.

We assume
$$
N \sim \mathrm{Bin}(10, p), \quad p = \frac{6}{10} = 0.6,
$$
so the pmf of $N$ is
$$
p_N(k) = \mathbb{P}(N = k)
       = \binom{10}{k} p^k (1-p)^{10-k},
\quad k = 0,1,\dots,10.
$$

Given $N = n$, the student will guess the remaining $10-n$ questions, each with success probability $1/2$. Thus
$$
Z \mid (N = n) \sim \mathrm{Bin}(10-n, 1/2),
$$
and the total score is
$$
Y = N + Z.
$$

We pass the student if $Y \ge T$, where $T \in \{0,1,\dots,10\}$.

---

### 1. Compute $\mathbb{P}(N < 5 \mid Y \ge T)$ for each $T$

We want, for each threshold $T$,
$$
\mathbb{P}(N < 5 \mid Y \ge T)
=
\frac{\mathbb{P}(N < 5,\, Y \ge T)}{\mathbb{P}(Y \ge T)}.
$$

We first compute the denominator using the law of total probability over $N$:
$$
\mathbb{P}(Y \ge T)
=
\sum_{n=0}^{10} \mathbb{P}(Y \ge T \mid N = n)\,\mathbb{P}(N = n).
$$

Given $N = n$, we have $Y = n + Z$ with $Z \sim \mathrm{Bin}(10-n, 1/2)$. Then
$$
Y \ge T
\quad \Longleftrightarrow \quad
n + Z \ge T
\quad \Longleftrightarrow \quad
Z \ge T - n.
$$

Therefore
$$
\mathbb{P}(Y \ge T \mid N = n)
=
\mathbb{P}(Z \ge T-n \mid N = n)
=
\sum_{z = T-n}^{10-n}
\binom{10-n}{z} \left(\frac12\right)^{10-n},
$$
with the edge cases:
- if $T-n \le 0$, then $\mathbb{P}(Y \ge T \mid N = n) = 1$,
- if $T-n > 10-n$, then $\mathbb{P}(Y \ge T \mid N = n) = 0$.

Thus
$$
\mathbb{P}(Y \ge T)
=
\sum_{n=0}^{10}
\mathbb{P}(Y \ge T \mid N = n)\, p_N(n).
$$

For the numerator we restrict to $N < 5$:
$$
\mathbb{P}(N < 5, Y \ge T)
=
\sum_{n=0}^{4}
\mathbb{P}(Y \ge T \mid N=n)\,\mathbb{P}(N=n)
=
\sum_{n=0}^{4}
\mathbb{P}(Y \ge T \mid N=n)\,p_N(n).
$$

Combining numerator and denominator we obtain
$$
\mathbb{P}(N < 5 \mid Y \ge T)
=
\frac{
\displaystyle \sum_{n=0}^{4}
\mathbb{P}(Y \ge T \mid N=n)\,p_N(n)
}{
\displaystyle \sum_{n=0}^{10}
\mathbb{P}(Y \ge T \mid N=n)\,p_N(n)
}.
$$

In the code, for each $T = 0,1,\dots,10$, we
1. loop over $n = 0,\dots,10$,
2. compute $p_N(n)$,
3. compute the tail probability
   $$
   \mathbb{P}(Z \ge T-n) = \sum_{z = T-n}^{10-n}
   \binom{10-n}{z} \left(\frac12\right)^{10-n},
   $$
   including the edge cases $T-n \le 0$ and $T-n > 10-n$,
4. accumulate the denominator over all $n$,
5. accumulate the numerator only over $n < 5$,
6. divide numerator by denominator.

We store these 11 values in the list `problem11_probabilities`, one value for each threshold $T$.

---

### 2. Smallest $T$ with 90% certainty that $N \ge 5$

We want the smallest threshold $T$ such that
$$
\mathbb{P}(N \ge 5 \mid Y \ge T) \ge 0.9.
$$

Using complements,
$$
\mathbb{P}(N \ge 5 \mid Y \ge T)
=
1 - \mathbb{P}(N < 5 \mid Y \ge T),
$$
so the condition is equivalent to
$$
\mathbb{P}(N < 5 \mid Y \ge T) \le 0.1.
$$

Using the values computed in part 1, we check for which $T$ the quantity
$\mathbb{P}(N < 5 \mid Y \ge T)$ is at most $0.1$.  

Numerically, this first happens at
$
T^* = 8,
$
for which
$$
\mathbb{P}(N < 5 \mid Y \ge 8) \approx 0.073
\quad \Rightarrow \quad
\mathbb{P}(N \ge 5 \mid Y \ge 8) \approx 0.927.
$$

Thus the smallest threshold that gives at least $90\%$ certainty that the
student knows at least 5 answers is $T^* = 8$.


````python
# Hint the PMF of N is p_N(k) where p_N is

from scipy.special import binom as binomial
p = 6/10
p_N = lambda k: binomial(10,k)*((1-p)**(10-k))*((p)**k)

````

````python
# Part 1: 
# replace XXX to represent P(N < 5) for T = [0,1,2,...,10], i.e. your answer should be a list
# of length 11.
from scipy.special import binom as binomial


def binom_tail_half(trials, k):
    """P(Binom(trials, 0.5) >= k) with simple edge cases."""
    if k <= 0:
        return 1.0
    if k > trials:
        return 0.0
    # sum_{z=k}^{trials} C(trials, z) * (1/2)^trials
    return sum(binomial(trials, z) for z in range(k, trials+1)) * (0.5 ** trials)

problem11_probabilities = []
for T in range(11):  # T = 0..10
    num = 0.0  # P(N<5, Y>=T)
    den = 0.0  # P(Y>=T)
    for n in range(11):  # n = 0..10
        py_ge_T_given_n = binom_tail_half(10-n, T-n)  # P(Z >= T-n)
        term = p_N(n) * py_ge_T_given_n
        den += term
        if n < 5:
            num += term
    problem11_probabilities.append(num / den)

problem11_probabilities  # list for T=0..10
    
for num in range(len(problem11_probabilities)):
    print(problem11_probabilities[num])
````

## Concentration of Measure

We examine which empirical quantities $Z$ satisfy **exponential** concentration
$$
P(Z - \mathbb{E}[Z] \ge \epsilon)
\le
C_1 e^{-C_2 n \epsilon^2}
\;\;\vee\;\;
C_3 e^{-C_4 n(\epsilon+1)}.
$$

We also check which quantities satisfy the weaker **polynomial** concentration
$$
P(Z - \mathbb{E}[Z] \ge \epsilon)
\le
\frac{C_1}{n\epsilon^2}.
$$

---

### 1. Exponential Concentration

We use the following principles:

- The empirical mean of sub-Gaussian variables is sub-Gaussian.
- The empirical mean of sub-Exponential variables is sub-Exponential.
- The empirical variance of sub-Gaussian variables concentrates exponentially because $X_i^2$ is sub-Exponential.
- Deterministic variables concentrate trivially.
- Bounded random variables (e.g. Bernoulli) satisfy Hoeffding's inequality.
- Finite-variance variables alone do *not* give exponential concentration.

From these rules, the following quantities concentrate **exponentially**:

- (2) Empirical variance of sub-Gaussian variables  
- (4) Empirical mean of sub-Gaussian variables  
- (5) Empirical mean of sub-Exponential variables  
- (9) Empirical mean of deterministic variables  
- (10) Empirical tenth moment of Bernoulli variables  

Thus:
$$
\text{Exponential concentration} = [2, 4, 5, 9, 10].
$$

---

### 2. Polynomial (Weak) Concentration

A quantity satisfies the weaker bound
$$
P(Z - \mathbb{E}[Z] \ge \epsilon)
\le
\frac{C_1}{n\epsilon^2}
$$
whenever Chebyshev's inequality applies, i.e. whenever the relevant variance exists.

This includes:

- (2) Empirical variance of sub-Gaussian variables  
- (3) Empirical variance of sub-Exponential variables  
- (4) Empirical mean of sub-Gaussian variables  
- (5) Empirical mean of sub-Exponential variables  
- (6) Empirical mean with finite variance  
- (7) Empirical third moment with finite sixth moment  
- (8) Empirical fourth moment of sub-Gaussian variables  
- (9) Empirical mean of deterministic variables  
- (10) Empirical tenth moment of Bernoulli variables  

Thus:
$$
\text{Weak concentration} = [2, 3, 4, 5, 6, 7, 8, 9, 10].
$$
