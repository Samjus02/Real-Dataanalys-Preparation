## 1. [1.5p]

We define the transition matrix

$$
P =
\begin{pmatrix}
0.3 & 0.4 & 0.3 \\
0.2 & 0.5 & 0.3 \\
0.4 & 0.3 & 0.3
\end{pmatrix}
$$

The probability of being in the downtown region after two steps, starting in the suburbs, is the entry in row 2 and column 1 of the squared matrix:

$$
P^2 = P \cdot P
$$

The entry is computed as

$$
(P^2)_{2,1}
= P_{2,1} P_{1,1}
+ P_{2,2} P_{2,1}
+ P_{2,3} P_{3,1}
$$

$$
= 0.2 \cdot 0.3
+ 0.5 \cdot 0.2
+ 0.3 \cdot 0.4
$$

$$
= 0.06 + 0.10 + 0.12 = 0.28
$$

Thus, the probability that the truck is in the downtown region after two time steps is

$$
0.28
$$


Or simply. Just multiply the matrixes and look ath the $P^2$ matrix from suburbs to downtown, and that's the correct value.

P =

A stationary distribution is a vector $\pi = (\pi_1, \pi_2, \pi_3)$ such that

$$
\pi P = \pi
$$

and

$$
\pi_1 + \pi_2 + \pi_3 = 1.

$$
\pi_1 = 0.3\pi_1 + 0.2\pi_2 + 0.4\pi_3

$$

$$
\pi_3 = 0.3\pi_1 + 0.3\pi_2 + 0.3\pi_3
$$

together with the normalization constraint

$$
\pi_1 + \pi_2 + \pi_3 = 1.
$$

````python
def stationary_distribution(P):
    """
    Computes the stationary distribution of a Markov chain
    by finding the eigenvector corresponding to eigenvalue 1.
    """
    eigenvalues, eigenvectors = np.linalg.eig(P.T)
    
    # Find the eigenvector associated with eigenvalue 1
    idx = np.argmin(np.abs(eigenvalues - 1))
    vec = np.real(eigenvectors[:, idx])
    
    # Normalize to sum to 1
    stationary = vec / np.sum(vec)
    return stationary

P = np.array([[0.3, 0.4, 0.3], [0.2, 0.5, 0.3], [0.4, 0.3, 0.3]])

print(stationary_distribution(P))
````

## 5. [2p] Expected hitting time to downtown

We consider a Markov chain with three regions: downtown (D), suburbs (S), and countryside (C), with transition matrix

$$
P = \begin{pmatrix}
0.3 & 0.4 & 0.3 \\\\
0.2 & 0.5 & 0.3 \\\\
0.4 & 0.3 & 0.3
\end{pmatrix}
$$

We want the expected number of steps until the chain first enters downtown when starting from suburbs. Define the hitting time

$$
T = \min\{n \ge 0 : X_n \text{ is in downtown}\},
$$

and let

$$
h(i) = \mathbb{E}[T \mid X_0 = i].
$$

Clearly,

$$
h_D = 0.
$$

For the non-target states S and C, the first-step recursion gives

$$
h(i) = 1 + \sum_j P_{ij} h(j).
$$

From the suburbs row of \(P\):

$$
h_S = 1 + 0.2 h_D + 0.5 h_S + 0.3 h_C,
$$

and from the countryside row:

$$
h_C = 1 + 0.4 h_D + 0.3 h_S + 0.3 h_C.
$$

Using $( h_D = 0 )$, these become

$$
h_S = 1 + 0.5 h_S + 0.3 h_C,
$$

$$
h_C = 1 + 0.3 h_S + 0.3 h_C.
$$

Rewriting:

$$
0.5 h_S - 0.3 h_C = 1,
$$

$$
-0.3 h_S + 0.7 h_C = 1.
$$

Solving yields

$$
h_C = \frac{40}{13}, \qquad h_S = \frac{50}{13}.
$$

Thus,

$$
\mathbb{E}[T \mid X_0 = S] = \frac{50}{13} \approx 3.8.
$$

### 4.1 [2p] Linear Congruential Generator

We implement a Linear Congruential Generator (LCG) as in the lecture notes (Definition 6.2).

An LCG with parameters $(a, b, M)$ on
$$
\{0, 1, \dots, M-1\}
$$
is defined by
$$
u_{k+1} = (a u_k + b) \bmod M,
$$
where $u_0$ is the seed. This produces a sequence $u_0, u_1, u_2, \dots$ of integers in $\{0,\dots,M-1\}$.

To obtain approximately Uniform$(0,1)$ random numbers we later scale by dividing by $M$. In this problem, however, the generator in part (1) is required to return integers in $\{0,\dots,\text{period}-1\}$, and part (2) performs the scaling.

We choose a large modulus
$$
M = 2^{32},
$$
and the parameters
$$
a = 1664525, \qquad b = 1013904223,
$$
which are the well-known "Numerical Recipes" LCG parameters. They satisfy the Hull–Dobell conditions (Theorem 6.8 in the notes) for full period $M$:

- $\gcd(b, M) = 1$, since $M$ is a power of $2$ and $b$ is odd.
- The only prime divisor of $M$ is $2$, and $2 \mid (a-1)$ since $a$ is odd.
- Because $4 \mid M$, we require $4 \mid (a-1)$; here $a-1 = 1664524$ is divisible by $4$.

Hence the LCG has period $M$.

The implementation is:

````python
import math

def problem4_accept_reject(uniformGenerator=None, n_iterations=None, seed=0):
    """
    Accept–Reject sampler to draw samples from
        p0(x) = (pi/2) * |sin(2*pi*x)|  for x in [0,1],
    using q(x) = Uniform(0,1) as the proposal.

    Parameters
    ----------
    uniformGenerator : callable
        A function with signature uniformGenerator(size, seed) that returns
        `size` IID samples in [0,1).
    n_iterations : int
        Number of proposal/acceptance attempts (uses 2 uniforms per iteration).
    seed : int
        Seed forwarded to `uniformGenerator`.

    Returns
    -------
    list[float]
        Accepted samples distributed as p0(x).
    """
    if uniformGenerator is None or n_iterations is None:
        raise ValueError("Provide a uniformGenerator(size, seed) and n_iterations.")

    # Draw 2 uniforms per iteration: one for proposal x, one for acceptance u
    uniforms = uniformGenerator(size=2 * n_iterations, seed=seed)

    accepted = []
    for i in range(0, 2 * n_iterations, 2):
        x = uniforms[i]
        u = uniforms[i + 1]

        # Be robust if the provided generator ever returns values outside [0,1)
        if not (0.0 <= x < 1.0):
            x = x % 1.0
        if not (0.0 <= u < 1.0):
            u = u % 1.0

        # With proposal q(x)=1 on [0,1] and M = max p0(x) = pi/2,
        # acceptance probability is p0(x) / (M*q(x)) = |sin(2*pi*x)|.
        if u <= abs(math.sin(2.0 * math.pi * x)):
            accepted.append(x)

    return accepted
````
